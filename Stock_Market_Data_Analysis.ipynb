**Risk Analysis**

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define Tata stocks
tata_stocks = ["TCS.NS", "TATAMOTORS.NS", "TATASTEEL.NS", "TITAN.NS", "TATACONSUM.NS", "TATAPOWER.NS"]

# Fetch historical stock prices (Last 2 Years)
data = yf.download(tata_stocks, period="2y")["Close"]

# Calculate Daily Returns
daily_returns = data.pct_change().dropna()

# Compute Risk Metrics
expected_return = daily_returns.mean()  # Mean of daily returns
volatility = daily_returns.std()  # Standard deviation (volatility)

# Create a Risk-Return DataFrame
risk_return_df = pd.DataFrame({"Expected Return": expected_return, "Volatility (Risk)": volatility})

# 📊 Scatter Plot: Risk vs. Return
plt.figure(figsize=(10, 6))
sns.scatterplot(x=risk_return_df["Volatility (Risk)"], y=risk_return_df["Expected Return"], s=100, color="darkblue")

# Annotate each stock
for stock in tata_stocks:
    plt.annotate(stock,
                 (risk_return_df.loc[stock, "Volatility (Risk)"], risk_return_df.loc[stock, "Expected Return"]),
                 textcoords="offset points", xytext=(5,5), ha="right", fontsize=10)

# Labels & Title
plt.xlabel("Volatility (Standard Deviation of Daily Returns)")
plt.ylabel("Expected Return (Mean of Daily Returns)")
plt.title("Risk vs. Return Analysis for Tata Stocks", fontsize=14)
plt.grid(True)

plt.show()
```

1️⃣**Historical (Bootstrap) VaR Based on Price**


*   Uses past price movements to estimate potential future losses.
*   We calculate the percentile of absolute price changes.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def compute_rsi(data, window=14):
    """Calculate the Relative Strength Index (RSI)."""
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Initialize an empty list for storing stock metrics
stock_metrics = []

for stock in tata_stocks:
    stock_data = tata_data['Close'][stock]

    # Compute Metrics
    mean_return = stock_data.pct_change().mean()
    volatility = stock_data.pct_change().std()
    correlation_nifty = stock_data.corr(nifty)
    rsi_last = compute_rsi(stock_data).iloc[-1]  # Latest RSI value

    # Append to list
    stock_metrics.append([stock, mean_return, volatility, correlation_nifty, rsi_last])

# Convert to DataFrame
columns = ["Stock", "Mean Return", "Volatility", "Correlation with NIFTY 50", "RSI"]
comparison_df = pd.DataFrame(stock_metrics, columns=columns)

#  Style Function for RSI Highlighting
def highlight_rsi(val):
    """Color-code RSI values: Red (Overbought), Green (Oversold), White (Neutral)."""
    color = "white"  # Default
    if val > 70:
        color = "red"
    elif val < 30:
        color = "green"
    return f"background-color: {color};"

# Apply Styling
styled_df = (comparison_df.style
             .applymap(highlight_rsi, subset=["RSI"])
             .set_properties(**{'background-color': 'black', 'color': 'white', 'border-color': 'gray'}))

# Display the styled table
styled_df
```

**Automated Trading Signals Based on RSI**
* We will generate Buy, Sell, and Hold signals using RSI levels:

**📌 Trading Strategy Based on RSI:**
* ✅ BUY 🟢 → RSI < 30 (Oversold) → Stock may be undervalued.
* ✅ SELL 🔴 → RSI > 70 (Overbought) → Stock may be overvalued.
* ✅ HOLD ⚪ → RSI between 30-70 → No strong signal.

```python
def generate_trading_signal(rsi_value):
    """Generate trading signals based on RSI."""
    if rsi_value < 30:
        return "BUY 🟢"
    elif rsi_value > 70:
        return "SELL 🔴"
    else:
        return "HOLD ⚪"

# Initialize a list for storing stock metrics
stock_metrics = []

for stock in tata_stocks:
    stock_data = tata_data['Close'][stock]

    # Compute Metrics
    mean_return = stock_data.pct_change().mean()
    volatility = stock_data.pct_change().std()
    correlation_nifty = stock_data.corr(nifty)
    rsi_last = compute_rsi(stock_data).iloc[-1]  # Latest RSI value
    trading_signal = generate_trading_signal(rsi_last)  # Generate Trading Signal

    # Append to list
    stock_metrics.append([stock, mean_return, volatility, correlation_nifty, rsi_last, trading_signal])

# Convert to DataFrame
columns = ["Stock", "Mean Return", "Volatility", "Correlation with NIFTY 50", "RSI", "Trading Signal"]
comparison_df = pd.DataFrame(stock_metrics, columns=columns)

#  Style Function for RSI Highlighting
def highlight_rsi(val):
    """Color-code RSI values: Red (Overbought), Green (Oversold), White (Neutral)."""
    color = "white"
    if val > 70:
        color = "red"
    elif val < 30:
        color = "green"
    return f"background-color: {color};"

# Apply Styling
styled_df = (comparison_df.style
             .applymap(highlight_rsi, subset=["RSI"])
             .set_properties(**{'background-color': 'black', 'color': 'white', 'border-color': 'gray'}))

# Display the styled table with Trading Signals
styled_df
```
